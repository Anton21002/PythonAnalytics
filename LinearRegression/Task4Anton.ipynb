For the independent values I choose Displacement (airflow) and horsepower .
Use cases:
Displacement could tell me that the more air moved with less fuel consumer the more cost-efficient the process ,so the more negative the correlation the better
The more horsepower the vehicle has the more fuel it consumes , so we can also add weight carried into the equasion to see what the most cost /fuel efficient vehicle to carry different weights. More horsepower- more weight , but more the consumption , so it could be like a desicion between running two light trucks or one massive truck to carry the same weight.

```python
sns.jointplot(data=cardata, x='Displacement100ci', y='GallonsPer100Miles')
sns.jointplot(data=cardata, x='Horsepower100', y='GallonsPer100Miles')
```

Gallons/Airflow - There are present models that have worse or better air , possible due to age of the vehicle.Overall less air capacity less fuel spent, maybe (im not a car expert) higher capacity for air is used in more horsepower hungry machines.
Speaking about horsepower - It is similar to the airflow , the more horsepower the vehicle has the more fuel it consumes,with maybe a little bit more variation

```python
X = cardata['Displacement100ci'].to_frame()
y = cardata['GallonsPer100Miles']
car_model = LinearRegression().fit(X, y)
car_model.coef_, car_model.intercept_
```

```python
car_model.predict(pd.DataFrame([3]))
car_model.predict(pd.DataFrame([1]))
```

```python
plt.scatter(car_model.predict(X), car_model.predict(X) - y)
plt.hlines(y=0, xmin=2, xmax=9)
```

The prediction is generally on point for airflow/fuel ,values are more-less around the y=0 ,the outliers in the predictions are similar to the same outliers in the original table.

```python
car_model.score(X, y)            
```

75% accuracy does mean that there is strong correlation between fuel consumtion and airflow .

```python
y =  cardata['GallonsPer100Miles']
X1 = cardata[['Cylinders', 'Displacement100ci', 'Horsepower100', 'Weight1000lb', 'Seconds0to60']]
X2 = cardata[['Displacement100ci', 'Horsepower100', 'Weight1000lb']]
X3 = cardata[['Displacement100ci','Horsepower100']]
```

```python
car_model1 = LinearRegression().fit(X1,y)
car_model2 = LinearRegression().fit(X2,y)
car_model3 = LinearRegression().fit(X3,y)
```

```python
print(car_model1.score(X1,y))
print(car_model2.score(X2,y))
print(car_model3.score(X3,y))
```

Increasing the amount variables does seem to increase the accuracy of predictions.Weight seems to also influence quite heavily the accuracy.

Now We will also check out the correlation between horsepower and fuel consumption

```python
Xh = cardata['Horsepower100'].to_frame() 
yh = cardata['GallonsPer100Miles']
car_horse = LinearRegression().fit(Xh, yh)
car_horse.coef_, car_horse.intercept_
```

```python
car_horse.predict(pd.DataFrame([3]))
car_horse.predict(pd.DataFrame([1]))
```

```python
plt.scatter(car_horse.predict(Xh), car_horse.predict(Xh) - yh)
plt.hlines(y=0, xmin=2, xmax=9)
```

Just like the other scatter graph ,the predictions are on-par with original values,with as usual having similar amount of outliers.

```python
car_horse.score(Xh, yh)
```

73% score tells that there is a strong correlation between horsepower and fuel consumption .
